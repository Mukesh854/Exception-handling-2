Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.


ArithmeticError is a base class for arithmetic-related errors in Python. Two common errors under this class are ZeroDivisionError, which occurs when dividing by zero, and OverflowError, which happens when a value exceeds the maximum representable value for its data type.

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is used for exceptions related to failed lookups in data structures. Two common subclasses are KeyError, which occurs when accessing a non-existent key in a dictionary, and IndexError, which happens when accessing an out-of-range index in a sequence like a list.






Q5. Explain ImportError. What is ModuleNotFoundError?


ImportError occurs when Python fails to import a module, typically due to reasons like missing module, incorrect path, or syntax errors. ModuleNotFoundError is a specific type of ImportError raised when the requested module cannot be found.
